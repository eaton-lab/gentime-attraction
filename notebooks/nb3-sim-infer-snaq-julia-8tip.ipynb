```julia
# load Vary Ne only simulated genealogies = input trees
unlinked_vary_gen_input = less("./unlinked_vary_Ne_input.tre")
```

```julia
unlinked_vary_Ne_input_mtrees = readMultiTopology("./unlinked_vary_Ne_input.tre")
```

```julia
# read all trees, calculate quartet CFs
q_411,t_411 = countquartetsintrees(unlinked_vary_Ne_input_mtrees);
df_411 = writeTableCF(q_411,t_411);
# save table
using CSV
CSV.write("tableCF_unlinked_vary_Ne_411.csv", df_411);

df_411
```

```julia
# create tableCF object
unlinked_var_gen_411_CF = readTableCF("tableCF_unlinked_vary_Ne_411.csv");
```

```julia
# load starting tree topologies for network estimation
tre1 = readMultiTopology("./snaq_starting_topology.tre")[1];
```

```julia
# network estimation with 0 to 3 hybrid edges
net0_411 = snaq!(tre1,unlinked_var_gen_411_CF, hmax=0, filename="net0_411", seed=1234)
net1_411 = snaq!(net0_411, unlinked_var_gen_411_CF, hmax=1, filename="net1_411", seed=2345)
net2_411 = snaq!(net1_411,unlinked_var_gen_411_CF, hmax=2, filename="net2_411", seed=3456)
net3_411 = snaq!(net0_411,unlinked_var_gen_411_CF, hmax=3, filename="net3_411", seed=4567)
```

```julia
# optimize number of hybrid edges = 1
scores_411 = [net0_411.loglik, net1_411.loglik, net2_411.loglik, net3_411.loglik]
using Gadfly;
plot(x=collect(0:3), y=scores_411, Geom.point, Geom.line)
```
